```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
df = sns.load_dataset("flights")
df.head()
```

```python
sns.relplot(x="year", y="passengers", data=df, kind="line", hue="month")
plt.show()
```

```python
sns.lineplot(x="year", y="passengers", data=df)
plt.show()
```
