```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
tips = sns.load_dataset("tips")
tips['tip_pct'] = 100 * tips['tip'] / tips['total_bill']
```

```python
grid = sns.FacetGrid(tips, row="sex", col="time", margin_titles=True)
grid.map_dataframe(sns.histplot, x="tip_pct", bins=np.linspace(0, 40, 15), color="steelblue")
grid.set_axis_labels("Tip Percentage", "Count")
plt.subplots_adjust(top=0.9)
grid.fig.suptitle("Distribution of Tip % by Sex and Time")
plt.show()
```
