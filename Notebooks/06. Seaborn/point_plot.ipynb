```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme()
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
sns.catplot(x='sex', y='survived', hue='class', kind='point', data=df)
plt.show()
```

```python
sns.catplot(
    x='class',
    y='survived',
    hue='sex',
    palette={'male': 'g', 'female': 'm'},
    markers=['^', 'o'],
    linestyles=['-', '--'],
    kind='point',
    data=df
)
plt.show()
```
