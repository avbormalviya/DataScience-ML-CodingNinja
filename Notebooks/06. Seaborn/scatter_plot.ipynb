```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
rng = np.random.RandomState()

x = np.linspace(0, 100, 100)
y = np.cumsum(rng.randn(100, 4), 0)
```

```python
plt.plot(x, y)
plt.legend('ABCD', ncol=2, loc='upper left')
```
