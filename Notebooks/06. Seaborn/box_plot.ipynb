```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
sns.catplot(x='sex', y='total_bill', data=df, kind='box', hue='smoker')
plt.show()
```

```python
sns.catplot(x='sex', y='total_bill', data=df, kind='boxen', hue='smoker')
plt.show()
```
