```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
sns.set_theme()
```

```python
data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
data = pd.DataFrame(data, columns=['x', 'y'])
```

```python
sns.histplot(data=data, x="x", color="skyblue", kde=True, stat="density", label="x", element="step")
sns.histplot(data=data, x="y", color="orange", kde=True, stat="density", label="y", element="step")
plt.title("Seaborn Histograms with KDE")
plt.legend()
plt.show()
```

```python
sns.histplot(data['x'], kde=True, color='skyblue', stat="density", label='x')
sns.histplot(data['y'], kde=True, color='orange', stat="density", label='y')
plt.title("Histogram with KDE for x and y")
plt.legend()
plt.show()
```
