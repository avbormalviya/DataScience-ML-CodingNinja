```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
sns.catplot(x='sex', y='survived', hue='class', kind='bar', data=df)
plt.show()
```

```python
sns.catplot(x='deck', kind='count', palette='ch:.25', data=df, hue='survived')
plt.show()
```

```python
sns.catplot(x='deck', kind='count', palette='pastel', edgecolor='.6', data=df, hue='class')
plt.show()
```
